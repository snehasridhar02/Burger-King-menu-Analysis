# Q1.What is the average calorie content of Burger King menu items?

```python
avg_cal=bk['Calories'].mean()
print("The average calorie content of burger king menu items is: ",avg_cal)
```

# Q2. Which menu category has the highest average fat content?

```python
bk['Category'].unique()
```

```python
import matplotlib.pyplot as plt

plt.figure (figsize=(5, 5)) 
menu_category = bk.groupby('Category')['Fat (g)'].mean()
explode = (0, 0.05, 0)
label=menu_category.index
plt.pie(menu_category,explode=explode,labels=label,autopct='%0.1f%%',shadow=True)
plt.title(" Average Percentage of Fat")
plt.show()
```

Burgers have the highest average fat content.

# Q3. Is there a correlation between the amount of saturated fat and cholesterol in Burger King menu items?

```python
import seaborn as sns

corr=bk.corr()
plt.figure(figsize=(9, 5))  
sns.heatmap(corr,annot=True)
plt.show()
```

```python
bk['Saturated Fat (g)'].corr(bk['Cholesterol (mg)'])
```

The correleation coeffecient of 0.76 indicates that there is a positive correleation between the amount of saturated fat and cholestrol in burger king menu items.

# Q7. Number of menu items per category?

```python
menu_items=bk.groupby('Category').size()
sns.barplot(x=menu_items.index, y=menu_items.values,palette=['Pink','Orange','Yellow'])
plt.title('Count of menu items per category')
plt.show()
```

From the above bar chart we can come to a conclusion that the breakfast category has more number of menu items that the other categories.

# Q8. What is the distribution of cholestrol and protein in burger king foods?

```python
plt.figure(figsize=(10, 5))  
bk_long = bk.melt(id_vars=['Category'], value_vars=['Cholesterol (mg)', 'Protein (g)'], var_name='Nutrient')

sns.barplot(x='Category', y='value', hue='Nutrient', data=bk_long, palette='Set1')

plt.title('Comparison of Cholesterol and Protein by Category')
plt.xlabel('Category')
plt.ylabel('Value')
plt.legend(title='Nutrient') 
plt.show()
```

Burgers have more cholestrol and protein
